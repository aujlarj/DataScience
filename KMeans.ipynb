# K-Means Clustering

This notebook shows why it is important to have domain experts and knowledge

<u>Example:</u>
If the domain experts can indicate that there are 4 type of clusters then it's easier to use this technique and predict them. Else we can see it looks like 2 blobs and can further cluters it to 'n' numbers

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Create Data/Clusters
#### Orginal Cluster = 4

```python
data = make_blobs(n_samples=200, n_features=2, 
                           centers=4, cluster_std=1.8,random_state=101)
```

### Visualized Original vs Predicted Clusters

#### Predicting = 2

```python
kmeans2 = KMeans(n_clusters=2)
kmeans2.fit(data[0])
print('C-Centers:\n',kmeans2.cluster_centers_)
# print(kmeans2.labels_)
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans2.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

#### Predicting = 3

```python
kmeans3 = KMeans(n_clusters=3)
kmeans3.fit(data[0])
print('C-Centers:\n',kmeans3.cluster_centers_)
# print(kmeans2.labels_)
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans3.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

#### Predicting = 4

```python
kmeans4 = KMeans(n_clusters=4)
kmeans4.fit(data[0])
print('C-Centers:\n',kmeans4.cluster_centers_)
# print(kmeans2.labels_)
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans4.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
kmeans5 = KMeans(n_clusters=5)
kmeans5.fit(data[0])
print('C-Centers:\n',kmeans5.cluster_centers_)
# print(kmeans2.labels_)
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans5.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
